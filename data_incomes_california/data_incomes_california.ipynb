```python
def extrat_columns_delete(df):
    container=['GEO_ID','NAME']
    container_id=[]
    delete_column=[]
    for index, row in df.iterrows():
        labels=re.search('!!([^!]+)!!',row['Label'])
        if labels:
            found = labels.group(1)
            if found not in container_id:
                container_id.append(found)
                container.append(row['Column Name'])
                print(row['Column Name'],row['Label'])
            else:
                delete_column.append(row['Label'])
    return container
```

```python
year_2015=extrat_columns_delete(df_col_2015)

year_2016=extrat_columns_delete(df_col_2016)

year_2017=extrat_columns_delete(df_col_2017)

year_2018=extrat_columns_delete(df_col_2018)

year_2019=extrat_columns_delete(df_col_2019)
year_2015
```
