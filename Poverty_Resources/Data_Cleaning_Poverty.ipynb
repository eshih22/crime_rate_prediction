```python
import pandas as pd
from pathlib import Path

poverty_2015 = pd.read_csv("Data_Resources_Yearly/Poverty_2015.csv")
poverty_2016 = pd.read_csv("Data_Resources_Yearly/Poverty_2016.csv")
poverty_2017 = pd.read_csv("Data_Resources_Yearly/Poverty_2017.csv")
poverty_2018 = pd.read_csv("Data_Resources_Yearly/Poverty_2018.csv")
poverty_2019 = pd.read_csv("Data_Resources_Yearly/Poverty_2019.csv")

poverty_2015.head()
```

```python
df = [
    (poverty_2015, "2015"),
    (poverty_2016, "2016"),
    (poverty_2017, "2017"),
    (poverty_2018, "2018"),
    (poverty_2019, "2019")]
```

```python

def clean(df):
    clean_df = []
    for dataframe, year in df:
            dataframe = dataframe.drop(columns=[column for column in dataframe.columns if column.endswith('M')])
            dataframe = dataframe.drop(columns=["GEO_ID"])
            dataframe = dataframe.drop(columns=dataframe.columns[dataframe.apply(lambda col: col.astype(str).str.contains('\(X\)').any())])
            dataframe["NAME"] = dataframe["NAME"].str.replace(', California|city|CDP|town', '', regex=True)
            dataframe["NAME"] = dataframe["NAME"].str.strip()
            dataframe = dataframe.drop(columns=dataframe.columns[(dataframe.astype(str) == 'N').all()])
            dataframe["Year"] = year   
            dataframe = dataframe[["Year", "NAME", "S1702_C02_001E", "S1702_C02_002E", "S1702_C02_015E",
                                   "S1702_C02_017E", "S1702_C02_020E", "S1702_C02_021E", "S1702_C02_022E",
                                   "S1702_C02_023E", "S1702_C02_024E", "S1702_C02_025E", "S1702_C02_026E",
                                   "S1702_C02_027E", "S1702_C02_028E", "S1702_C02_029E", "S1702_C02_030E",
                                   "S1702_C02_031E", "S1702_C02_032E", "S1702_C02_033E", "S1702_C02_034E",
                                   "S1702_C02_035E", "S1702_C02_036E", "S1702_C02_037E", "S1702_C02_038E",
                                   "S1702_C02_039E", "S1702_C02_041E", "S1702_C02_042E"]]
            dataframe = dataframe.rename(columns={
                "NAME": "City",
                "S1702_C02_001E": "% All Families",
                "S1702_C02_002E": "% With related children of householder under 18 years",
                "S1702_C02_015E": "% Householder worked",
                "S1702_C02_017E": "% Householder 65 years and over",
                "S1702_C02_020E": "% less than high school graduate",
                "S1702_C02_021E": "% High school graduate",
                "S1702_C02_022E": "% Some college, associate degree",
                "S1702_C02_023E": "% Bachelor's degree or higher",
                "S1702_C02_024E": "% No related children under 18",
                "S1702_C02_025E": "% 1-2 related children under 18 ",
                "S1702_C02_026E": "% 3-4 related children under 18",
                "S1702_C02_027E": "% 5 or more related children under 18",
                "S1702_C02_028E": "% No own children under 18",
                "S1702_C02_029E": "% 1-2 own children under 18 ",
                "S1702_C02_030E": "% 3-4 own children under 18",
                "S1702_C02_031E": "% 5 or more own children under 18",
                "S1702_C02_032E": "% Family size 2",
                "S1702_C02_033E": "% Family size 3-4",
                "S1702_C02_034E": "% Family size 5-6",
                "S1702_C02_035E": "% Family size more than 7",
                "S1702_C02_036E": "% No workers in Family",
                "S1702_C02_037E": "% 1 workers in Family",
                "S1702_C02_038E": "% 2 workers in Family",
                "S1702_C02_039E": "% 3 or more workers in Family",
                "S1702_C02_041E": "% Owner occupied",
                "S1702_C02_042E": "% Renter occupied"})
            dataframe = dataframe.drop(dataframe.index[0])
            clean_df.append(dataframe)
    return clean_df
cleaned_df = clean(df)
```

```python
cleaned_df[0]
```

```python
years = ["2015", "2016", "2017", "2018", "2019"]
file_name = {f"Cleaned_Poverty_{year}": df for df, year in zip(cleaned_df, years)}
for name, df in file_name.items():
    csv_path= f"{name}.csv"
    df.to_csv(csv_path, index=False)
    
```
