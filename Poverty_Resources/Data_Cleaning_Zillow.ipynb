```python
import pandas as pd
from pathlib import Path
import matplotlib.pyplot as plt
import numpy as np

house_price = pd.read_csv("House_price_zillow.csv")
house_price.head()
```

```python
years = range(2015,2020)
year_mean = []
for year in years:
    start = f"1/31/{year}"
    end = f"12/31/{year}"
    mean = house_price.loc[:, start:end].mean(axis=1)
    year_mean.append((year, round(mean,1)))

for year, mean in year_mean:
    house_price[str((year))] = mean
    
selected_columns = ["RegionName"] + [str(year) for year in years]
clear_house_price = house_price[selected_columns]

clear_house_price = clear_house_price.sort_values(by="RegionName")
clear_house_price = clear_house_price.reset_index(drop=True)
clear_house_price.dropna(how='any', inplace=True)
clear_house_price
```

```python
csv_path = f"Clean_House_Price.csv"
clear_house_price.to_csv(csv_path, index=False)
```
