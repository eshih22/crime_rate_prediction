```python
# merge the 5 csv

# create year column for each csv
df_2015 = pd.read_csv('Age_2015.csv')
df_2015['Year'] = 2015

df_2016 = pd.read_csv('Age_2016.csv')
df_2016['Year'] = 2016

df_2017 = pd.read_csv('Age_2017.csv')
df_2017['Year'] = 2017

df_2018 = pd.read_csv('Age_2018.csv')
df_2018['Year'] = 2018

df_2019 = pd.read_csv('Age_2019.csv')
df_2019['Year'] = 2019
```

```python
# coombine the csv's into one
merged_df = pd.concat([df_2015, df_2016, df_2017, df_2018, df_2019], ignore_index=True)
merged_df
```

```python
# create a new merged csv file
merged_df.to_csv('Merged_Age.csv', index=False)
```
